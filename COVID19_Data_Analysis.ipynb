```python
# STEP 1 Import Libraries and Load the Dataset
import pandas as pd  # For data manipulation
import seaborn as sns  # For data visualization
import matplotlib.pyplot as plt  # For plotting

# Load the COVID-19 dataset
from google.colab import files
uploaded = files.upload()  # Upload the CSV file

df = pd.read_csv('covid19_confirmed_global.csv')

# Preview the dataset
df.head()  # View the first few rows
```

```python
# Exploring the Dataset

# Check the structure of the dataset
df.info()

# Preview the column names
df.columns

# Check for missing values
df.isnull().sum()

# Get basic statistics of numerical columns
df.describe()
```

```python
# Print the column names to check if 'Lat' and 'Long' exist
print(df.columns)
# Check and drop the columns only if they are present
columns_to_drop = ['Lat', 'Long']
df = df.drop(columns=[col for col in columns_to_drop if col in df.columns], axis=1)

#RESHAPING
df_melted = df.melt(id_vars=['Province/State', 'Country/Region'],
                    var_name='Date',
                    value_name='Confirmed_Cases')

# Convert the 'Date' column to datetime format
df_melted['Date'] = pd.to_datetime(df_melted['Date'])

#Grouping
df_grouped = df_melted.groupby(['Country/Region', 'Date'])['Confirmed_Cases'].sum().reset_index()

# Preview the cleaned dataset
df_grouped.head()

```

```python
df_us = df_grouped[df_grouped['Country/Region'] == 'US']
```

```python
#VISUALIZATION USING SEABORN AND MATPLOTLIB

#LINE PLOT
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_us, x='Date', y='Confirmed_Cases')
plt.title('COVID-19 Confirmed Cases Over Time in the US')
plt.show()
```

```python
# BARPLOT
# Get the total confirmed cases by country
df_total_cases = df_grouped.groupby('Country/Region')['Confirmed_Cases'].max().reset_index()

# Sort by confirmed cases and get the top 10
df_top10 = df_total_cases.sort_values(by='Confirmed_Cases', ascending=False).head(10)

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=df_top10, x='Country/Region', y='Confirmed_Cases')
plt.title('Top 10 Countries with the Highest Confirmed Cases')
plt.xticks(rotation=45)
plt.show()
```

```python
# HEAT MAP
# Filter for specific countries (US, India, Brazil, Russia)

countries = ['US', 'India', 'Brazil', 'Russia']
df_countries = df_grouped[df_grouped['Country/Region'].isin(countries)]

# Pivot the data for the heatmap
df_pivot = df_countries.pivot(index='Date', columns='Country/Region', values='Confirmed_Cases')
# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df_pivot, cmap='coolwarm', linewidths=0.5)
plt.title('COVID-19 Spread Over Time in Selected Countries')
plt.show()

```
